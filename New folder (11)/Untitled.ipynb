```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("TMeURO.csv")
df
```

```python
df.info()
```

```python
df = df.drop(columns='Longtitude',axis=1)
```

```python
df = df.drop(columns='Latitude',axis=1)
```

```python
df
```

```python
df.rename(columns={'City Code': 'City'}, inplace=True)
df
```

```python
df.rename(columns={'Order Priority Code': 'Order Priority'}, inplace=True)
df
```

```python
duplicate = df[df.duplicated()]
duplicate.shape
```

```python
df[df.duplicated()]
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.isnull().sum()
```

```python
city =df['City'].median()
city
```

```python
order = df['Order Priority'].median()
order
```

```python
df['City'].replace(to_replace=np.nan,value=city,inplace=True)
df['Order Priority'].replace(to_replace=np.nan,value=order,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.head()
product_sales = df.groupby('product')['total'].sum().reset_index()

top_5_products = product_sales.sort_values(by='total', ascending=False).head(5)

print(top_5_products)


plt.figure(figsize=(10, 6))
sns.barplot(x='total', y='product', data=top_5_products, palette='viridis')
plt.title('Top 5 Best Selling Products')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product')
plt.show()
```

```python
payment_method_distribution = df['payment_method'].value_counts()
print(payment_method_distribution)

plt.figure(figsize=(8, 8))
plt.pie(payment_method_distribution, labels=payment_method_distribution.index, autopct='%1.1f%%', colors=['#FF9999','#66B3FF','#99FF99','#FFCC99','#FFD700'], startangle=140)
plt.title('Payment Method Distribution')

plt.axis('equal')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')


df['Month'] = df['date'].dt.month_name()


monthwise_sales = df.groupby('Month')['total'].sum().reset_index()


month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthwise_sales['Month'] = pd.Categorical(monthwise_sales['Month'], categories=month_order, ordered=True)
monthwise_sales = monthwise_sales.sort_values('Month')

print(monthwise_sales)

plt.figure(figsize=(10, 6))
plt.plot(monthwise_sales['Month'], monthwise_sales['total'], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

plt.title('Monthwise Sales', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (in units)', fontsize=12)

plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```

```python
city_mapping = {
    1: 'Karachi',
    2: 'Lahore',
    3: 'Islamabad',
    4: 'Faisalabad',
    5:  'Quetta',
    6:  'Peshawar'
}


df['City'] = df['City'].map(city_mapping)

citywise_sales = df.groupby('City')['total'].sum().reset_index()


citywise_sales = citywise_sales.sort_values(by='total', ascending=False)

print(citywise_sales)

plt.figure(figsize=(10, 6))
sns.barplot(x='total', y='City', data=citywise_sales, palette='viridis')

plt.title('City-wise Sales', fontsize=14)
plt.xlabel('Total Sales (in units)', fontsize=12)
plt.ylabel('City', fontsize=12)


plt.show()
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')


df['Month'] = df['date'].dt.month_name()

monthly_sales_category = df.groupby(['Month', 'category'])['total'].sum().unstack()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_sales_category = monthly_sales_category.reindex(month_order)

print(monthly_sales_category)

monthly_sales_category.plot(kind='bar', figsize=(12, 8), width=0.8, color=['black', 'grey', 'green'])


plt.title('Monthly Sales by Category', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (in units)', fontsize=12)

plt.xticks(rotation=45)

plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout() 
plt.show()
```
